```python
def answears(df, level=0):
    a = []
    for index, row in df.iterrows():
        incorrect = row['incorrect_answers'].copy()
        if level == 2:
            incorrect.pop(random.randint(0,2))
        if level == 3:
            incorrect.pop(random.randint(0,2))
            incorrect.pop(random.randint(0,1))
        correct = row['correct_answer']
        incorrect += [correct]
        print(incorrect)
        a.append(incorrect)
    return a
```

```python
def shuffle_answears(answears):
    for sublist in answears:
        random.shuffle(sublist)
    return answears
```

```python
def find_true(answers, true_answers):
    true_list = []
    for answer, true_answer in zip(answers, true_answers):
        try:
            true_list.append(answer.index(true_answer))
        except ValueError:
            print('Something went wrong... \n '
                  f"true answear {true_answer} in {answer} dont exist")
    return true_list
```

```python
a = answears(df)
```

```python
shuffle = shuffle_answears(a)
true_answers = df.correct_answer.tolist()
```

```python
true_answers_index = find_true(shuffle, true_answers)
```
